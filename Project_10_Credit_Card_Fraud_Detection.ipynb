#Credit Card Fraud Detection

##Importing the Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

##Data Collection and Preprocessing

```python
#importing dataset into pandas Datafame
dataset = pd.read_csv('/content/creditcard.csv')
```

```python
#printing the number of rows an columns
dataset.shape
```

```python
#printing the first 5 rows
dataset.head()
```

```python
#printing the last 5 rows
dataset.tail()
```

```python
#getting info about dataset
dataset.info()
```

```python
#distribution of llegit transactiions and fradulent transaction
dataset['Class'].value_counts()
```

This is highly unbalanced

```python
# sperating the data for analysis
legit = dataset[dataset.Class == 0]
fraud = dataset[dataset.Class == 1]
```

```python
print(legit.shape)
print(fraud.shape)
```

```python
#Statistical measures for the transaction
legit.Amount.describe()
```

```python
fraud.Amount.describe()
```

```python
#compare the values for both transaction
dataset.groupby('Class').mean()
```

## Under sample


Build a sample dataset containing similiar distribution of legit and fraud transaction

```python
#number of fraudulent transaction == 492
legit_sample = legit.sample(n=492)
```

##Concatenating two Dataframe

```python
new_dataset = pd.concat([legit_sample, fraud], axis=0)
```

```python
new_dataset.head()
```

```python
new_dataset.tail()
```

```python
new_dataset['Class'].value_counts()
```

##Splitting dataset into data target

```python
x = new_dataset.drop(columns='Class', axis=1)
y = new_dataset['Class']
```

```python
print(x)
```

```python
print(y)
```

##Splitting into training and test

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, stratify=y, random_state=2)
```

```python
print(x.shape, x_train.shape, x_test.shape)
```

##Training Model

```python
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression()
classifier.fit(x_train,y_train)
```

##Model Evaluation

```python
#accuracy score on training data
from sklearn.metrics import accuracy_score
x_train_prediction = classifier.predict(x_train)
training_data_accuracy = accuracy_score(x_train_prediction, y_train)
```

```python
print('Accuracy on training data : ', training_data_accuracy)
```

```python
#accuracy score on test data
from sklearn.metrics import accuracy_score
x_test_prediction = classifier.predict(x_test)
test_data_accuracy = accuracy_score(x_test_prediction, y_test)
```

```python
print('Accuracy on test data : ', test_data_accuracy)
```
